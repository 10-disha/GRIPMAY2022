### Importing libraries

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

### Reading data via CSV file

```python
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
s_data = pd.read_csv(url)
print("Data imported successfully")

s_data.head(11)
```

### Plotting the distribution of scores

```python
s_data.plot(x='Hours', y='Scores', style='o', color='blue')  
plt.title('Hours vs Percentage')  
plt.xlabel('Number of Hours Studied')  
plt.ylabel('Percentage')  
plt.show()
```

### Dividing the data into attributes and labels

```python
x = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values  
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

### Applying Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

### Training Data

```python
regressor = LinearRegression()  
regressor.fit(x_train, y_train) 

print("Model is trained successfully.")
```

### Plotting the regression line and test data 

```python
line = regressor.coef_*x+regressor.intercept_
plt.scatter(x, y, color='red')
plt.plot(x, line);
plt.show()
```

### Predicting the Model

```python
print(x_test)
y_pred = regressor.predict(x_test)
```

### Comparing Actual and Predicted data 

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

### Testing Data 

```python
hours = 9.25
answer = regressor.predict([[hours]])
print("Number of Hours = {}".format(hours))
print("Predicted Percentage = {}".format(answer[0]))
```

### Evaluating the model using Mean Square Error(MSE)

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
```
